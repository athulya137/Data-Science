#Precision: It is the ratio of correctly predicted positive observations to the total predicted positive observations. Precision focuses on the accuracy of the positive predictions made by the model.
precision=(TP)/(TP+FP)
Recall: It is the ratio of correctly predicted positive observations to all the actual positive observations. Recall focuses on capturing all the actual positives.
Recall=TP/(TP+FN)
Precision emphasizes minimizing false positives, while recall emphasizes minimizing false negatives. In certain cases (like fraud detection or medical diagnoses), high recall is prioritized to ensure no positive cases are missed, whereas in other cases (like spam detection), precision is prioritized to reduce false alarms.

#Cross-Validation: It is a technique used to assess how well a model generalizes to an independent dataset (i.e., unseen data). It divides the data into several subsets or "folds" and uses some for training and others for validation, rotating through all subsets. The most common type is k-fold cross-validation, where the dataset is split into k subsets (or folds), and the model is trained k times, each time using a different fold for validation and the remaining folds for training.
--reduce overfiting
--more reliable model evaluation
--optimizing hyperparameters
--imbalance handling
